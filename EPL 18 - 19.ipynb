# Which Team Has The Best Attack

```python
ATKTeams = data[['attack_scored','attack_goals_headed','attack_goals_penalty','attack_goals_box','attack_goals_outsidebox'
                 ,'attack_goals_counter','attack_goals_freekick']]
Teams = data[['Team','category','general_league_position']]
ATKTeams.head()
```

```python
Best_ATK = scaler.fit_transform(ATKTeams)
Best_ATK = pd.DataFrame(Best_ATK, columns=['attack_scored','attack_goals_headed','attack_goals_penalty','attack_goals_box','attack_goals_outsidebox'
                 ,'attack_goals_counter','attack_goals_freekick'])

Best_ATK.reset_index(drop=True, inplace=True)
Teams.reset_index(drop=True, inplace=True)

Attacker = pd.concat([Teams, Best_ATK], axis=1)
Attacker
```

```python
Attacker['Total_Attack'] = (Attacker['attack_scored']+Attacker['attack_goals_headed']+Attacker['attack_goals_penalty']+
                            Attacker['attack_goals_box']+Attacker['attack_goals_outsidebox']+Attacker['attack_goals_counter']+
                            Attacker['attack_goals_freekick'])/7
Attacker.nlargest(1, 'Total_Attack')
```

# Which Team Has The Best Defend ?

```python
DEFTeams = data[['defence_saves','defence_blocks','defence_interceptions','defence_tackles',
                 'defence_tackles_last_man','defence_clearances','defence_clearances_headed']]
Teams = data[['Team','category','general_league_position']]
DEFTeams.head()
```

```python
Best_DEF = scaler.fit_transform(DEFTeams)
Best_DEF = pd.DataFrame(Best_DEF, columns=['defence_saves','defence_blocks','defence_interceptions','defence_tackles',
                 'defence_tackles_last_man','defence_clearances','defence_clearances_headed'])

Best_DEF.reset_index(drop=True, inplace=True)
Teams.reset_index(drop=True, inplace=True)

Defender = pd.concat([Teams, Best_DEF], axis=1)
Defender
```

```python
Defender['Total_Defend'] = (Defender['defence_saves']+Defender['defence_tackles']+Defender['defence_blocks']+
                            Defender['defence_clearances_headed']+Defender['defence_clearances']+Defender['defence_interceptions']+
                            Defender['defence_tackles_last_man'])/7

Defender.nlargest(1, 'Total_Defend')
```
