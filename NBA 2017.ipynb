# Preproccesing

```python
# Filter Season 2017 Only
data = df[df.Year==2017]
data.head()
```

```python
# Drop Column With Missing Values in all rows
data = data.drop(['blanl','blank2'], axis=1)
data.head()
```

```python
# Duplicate Values
data = data.drop_duplicates(['Player'])
data.head()
```

# Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

```python
# Get max and min age for each team
Aggre = {'Age': (('Oldest_Age','max'),('Youngest_Age','min'))}
Age = data.groupby(['Tm'], as_index=False).agg(Aggre)
Age
```

```python
# Get Oldest And Youngest Player in each team

Oldest_Player = []
Youngest_Player = []

for x in range(len(Age)):
    Oldest_Player.append(data.loc[(data["Age"]==Age[('Age', 'Oldest_Age')].values[x])
                                  & (data["Tm"]==Age["Tm"].values[x]), "Player"].values[0])
    Youngest_Player.append(data.loc[(data["Age"]==Age[('Age', 'Youngest_Age')].values[x])
                                    & (data["Tm"]==Age["Tm"].values[x]),"Player"].values[0])

Age['Oldest_Player'], Age['Youngest_Player'] = Oldest_Player, Youngest_Player
Age
```

# Which player has the most minutes played (MP) in each position (Pos)?

```python
data['MaxMP'] = data.groupby('Pos', as_index=False)['MP'].rank(ascending=0)
MostMP = data[data['MaxMP']==1.0][['Player','MP', 'Pos', 'Tm']]
MostMP
```

# Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?

```python
HighestAVG = data.groupby('Tm', as_index=False)['Tm','TRB%','AST%','STL%','BLK%'].mean()
HighestAVG.nlargest(1, 'TRB%')
```

# Which team has the best average stat record of their players? Note: you can refer to points, assists, rebounds, or anything else. A combination of several variables would be nice?

```python
Player = data[['Player','Tm']]
Ability = data[['DRB%','TRB%','AST%','STL%','BLK%','TOV%','USG%','TS%','FG%','FT%']]
Player.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
AB_rate = scaler.fit_transform(Ability)
Best_Rate = pd.DataFrame(AB_rate, columns=['DRB%','TRB%','AST%','STL%','BLK%','TOV%','USG%','TS%','FG%','FT%'])

Best_Rate.reset_index(drop=True, inplace=True)
Player.reset_index(drop=True, inplace=True)

Rating = pd.concat([Player, Best_Rate], axis=1)
```

```python
Rating.head()
```

```python
Rating['rating'] = (Rating['DRB%']+Rating['TRB%']+Rating['AST%']+Rating['TS%']+Rating['STL%']+Rating['BLK%']
+Rating['FG%']+Rating['FT%']+Rating['TOV%']+Rating['USG%'])/10
Rating.nlargest(1,'rating')
```

# Which team has the best average stat record of their players? Note: you can refer to points,
assists, rebounds, or anything else. A combination of several variables would be nice

```python
best_team = Rating.groupby('Tm').mean()
best_team.sort_values('rating', ascending=0).head(1)
```
